### Check for missing values

```python
# count missing values

missing = application_df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

missing_percentage = (missing/ len(application_df))*100
missing_df = pd.DataFrame({"Missing Values": missing, "Percentage": missing_percentage})
missing_df
```

### Draw a graph showing the percentage off classes

```python
import matplotlib.pyplot as plt

# counts intancesof each class
target_counts = application_df["TARGET"].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(target_counts, labels=["Clients who repaid the loan (0)", "unpaid loan (1)"], autopct="%1.1f%%", startangle=90, colors=["Blue", "Red"])
plt.axis("equal") # Ensures the pie chart is circular
plt.title("Distribution of Loan Repayment Status")
plt.legend(loc=1)
plt.show()
```

The results shows that 91.9% of applicants repaid their loans, while 8.1% defaulted.   
**What Does This Mean?**
*  This confirms that loan default are relatively rare.
*  Since the dataset is imbalanced, meaning there are far fewer cases of default, prediction models might struggle to learn patterns for unpaid loans.
*  A typical lender would expect defaults to be much lower than repaid loans, which aligns with real-world lending behavior.

## Problem 4: Data Exploration

### Issue 1: Do income levels affect loan default?

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=application_df[application_df['TARGET'] == 0], x='AMT_INCOME_TOTAL', label='No default')
sns.kdeplot(data=application_df[application_df['TARGET'] == 1], x='AMT_INCOME_TOTAL', label='Default')
plt.title('Income Distribution by Loan Default Status')
plt.xlim(-1e6, 2e6)
plt.legend()
plt.show()
```

### Issue 2: Does the loan amount or credit amount relate to the likelihood of default?

```python
sns.boxplot(x='TARGET', y='AMT_CREDIT', data=application_df)
plt.title('Loan Credit Amount vs. Default')
plt.xticks([0, 1], ['No Default', 'Default'])
plt.show()
```

### Issue 3: Are certain occupations or type of employment riskier than others?

```python
occ_df = application_df[['OCCUPATION_TYPE', 'TARGET']].groupby('OCCUPATION_TYPE').mean().sort_values(by='TARGET', ascending=False)
occ_df.plot(kind='bar', legend=False)
plt.ylabel('Default Rate')
plt.title('Default Rate by Occupation')
plt.xticks(rotation=90)
plt.show()
```

### Issue 4: Does age play a role in the probability of default?

```python
application_df['AGE_YEARS'] = application_df['DAYS_BIRTH']/-365

sns.kdeplot(data=application_df[application_df["TARGET"]==0], x ='AGE_YEARS', label='No default')
sns.kdeplot(data=application_df[application_df["TARGET"]==1], x ='AGE_YEARS', label='Default')
plt.title('Age Distribution by Loan Default Status')
plt.legend()
plt.show()
```

### Issue 5: Is family status (e.g., number of children or martial status) linked to default rates?

```python
child_df = application_df.groupby('CNT_CHILDREN')['TARGET'].mean()

child_df.plot(kind='bar')
plt.ylabel('Default Rate')
plt.title('Default Rate by Number of Children')
plt.xlabel('Number of Children')
plt.show()
```

### Issue 6: Is family status (e.g., marital status) linked to default rates?

```python
marital_status_df = application_df.groupby('NAME_FAMILY_STATUS')['TARGET'].mean()

marital_status_df.plot(kind='bar')
plt.ylabel('Default Rate')
plt.title('Default Rate by Family Status')
plt.xlabel('Family Status')
plt.show()
```
